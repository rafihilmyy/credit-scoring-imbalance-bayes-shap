## **1.2 Sample Splitting**
---

```python
# Split response and predictors
y = data[response_variable]
X = data.drop(columns = [response_variable],
              axis = 1)

# Validate the splixtting
print('y shape :', y.shape)
print('X shape :', X.shape)
```

```python
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    stratify = y,
                                                    test_size = 0.3,
                                                    random_state = 42)

# Validate splitting
print('X train shape :', X_train.shape)
print('y train shape :', y_train.shape)
print('X test shape  :', X_test.shape)
print('y test shape  :', y_test.shape)
```

```python
y_train.value_counts(normalize = True)
```

```python
y_test.value_counts(normalize = True)
```

```python
# Concatenate X_train and y_train as data_train
data_train = pd.concat((X_train, y_train),
                       axis = 1)

# Validate data_train
print('Train data shape:', data_train.shape)
data_train.head()
```

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
# 1. Pisahkan kembali X_train dan y_train dari data_train
X_train = data_train.drop(columns='TARGET')
y_train = data_train['TARGET']
```

```python
# 2. Gabungkan kembali dengan X_test
X_full = pd.concat([X_train, X_test], axis=0)
```

```python
# 3. Encoding kategorikal
X_encoded = pd.get_dummies(X_full, columns=cat_columns, drop_first=True)
```

```python
# 4. Bagi kembali ke train dan test (berdasarkan ukuran asli)
X_train_encoded = X_encoded.iloc[:X_train.shape[0], :].copy()
X_test_encoded = X_encoded.iloc[X_train.shape[0]:, :].copy()
```

```python
from imblearn.under_sampling import RandomUnderSampler

# Buat objek RandomUnderSampler
rus = RandomUnderSampler(
        sampling_strategy="majority",
        random_state=42
      )

# Fit objek
X_train_encoded, y_train = rus.fit_resample(X_train_encoded, y_train)
```

```python
# # 5. Training model XGBoost
# xgb_model = XGBClassifier(
#     n_estimators=100,
#     max_depth=4,
#     learning_rate=0.1,
#     subsample=0.8,
#     colsample_bytree=0.8,
#     # scale_pos_weight=11.4,  # disesuaikan dengan imbalance rasio
#     use_label_encoder=False,
#     eval_metric='logloss',
#     random_state=42
# )
# asli sebelum opt
```

```python
# 5. Training model XGBoost
xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=10,
    learning_rate=0.018859918437244137,
    subsample=0.8558942015641932,
    colsample_bytree=0.808034848941382,
    # scale_pos_weight=11.4,  # disesuaikan dengan imbalance rasio
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)
# Best Params: OrderedDict({'colsample_bytree': 0.808034848941382, 'learning_rate': , 'max_depth': 10, 'n_estimators': 300, 'subsample': 0.8558942015641932})
```

```python
xgb_model.fit(X_train_encoded, y_train)
```

```python
# 6. Prediksi dan evaluasi
y_pred_test = xgb_model.predict(X_test_encoded)
y_proba_test = xgb_model.predict_proba(X_test_encoded)[:, 1]
```

```python
# Classification report
print(" Classification Report  XGBoost")
print(classification_report(y_test, y_pred_test))
```

```python
# AUC
print("AUC:", roc_auc_score(y_test, y_proba_test))
```

```python
# Confusion Matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(4, 3))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix  XGBoost")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, roc_auc_score
import numpy as np
import pandas as pd

# === Confusion Matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_test).ravel()

# === Metrics
auc = roc_auc_score(y_test, y_proba_test)
tpr = tp / (tp + fn)  # Sensitivity / Recall
tnr = tn / (tn + fp)  # Specificity
fpr = fp / (fp + tn)
gmean = np.sqrt(tpr * tnr)
f1 = f1_score(y_test, y_pred_test)
accuracy = accuracy_score(y_test, y_pred_test)

# === Buat DataFrame
xgb_metrics_df = pd.DataFrame([{
    'Classifier': 'XGBoost',
    'AUC': round(auc, 4),
    'TPR': round(tpr, 4),
    'TNR': round(tnr, 4),
    'FP - Rate': round(fpr, 4),
    'G - Mean': round(gmean, 4),
    'F1 - Score': round(f1, 4),
    'Akurasi': round(accuracy, 4)
}])

# Tampilkan
xgb_metrics_df
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, make_scorer
from skopt import BayesSearchCV
from skopt.space import Real, Integer
```

```python
# 1. Definisikan ruang hyperparameter
param_space = {
    'n_estimators': Integer(50, 300),
    'max_depth': Integer(3, 10),
    'learning_rate': Real(0.01, 0.3, prior='log-uniform'),
    'subsample': Real(0.5, 1.0),
    'colsample_bytree': Real(0.5, 1.0)
    # 'scale_pos_weight': Real(1, 20)
}
```

```python
# 2. Cross-validator
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# 3. BayesSearchCV setup
opt = BayesSearchCV(
    estimator=XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42
    ),
    search_spaces=param_space,
    scoring='roc_auc',
    cv=cv,
    n_iter=30,  # jumlah iterasi pencarian
    n_jobs=-1,
    verbose=0,
    random_state=42
)
```

```python
# 4. Fit ke data training
opt.fit(X_train_encoded, y_train)
```

```python
# 5. Model terbaik
best_model = opt.best_estimator_
print("Best Params:", opt.best_params_)
```

```python
# 6. Prediksi dan evaluasi
y_pred_test = best_model.predict(X_test_encoded)
y_proba_test = best_model.predict_proba(X_test_encoded)[:, 1]
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score

tn, fp, fn, tp = confusion_matrix(y_test, y_pred_test).ravel()

auc = roc_auc_score(y_test, y_proba_test)
tpr = tp / (tp + fn)
tnr = tn / (tn + fp)
fpr = fp / (fp + tn)
gmean = np.sqrt(tpr * tnr)
f1 = f1_score(y_test, y_pred_test)
accuracy = accuracy_score(y_test, y_pred_test)
```

```python
xgb_bayes_metrics_df = pd.DataFrame([{
    'Classifier': 'XGBoost + BayesOpt',
    'AUC': round(auc, 4),
    'TPR': round(tpr, 4),
    'TNR': round(tnr, 4),
    'FP - Rate': round(fpr, 4),
    'G - Mean': round(gmean, 4),
    'F1 - Score': round(f1, 4),
    'Akurasi': round(accuracy, 4)
}])
```
