```python
from scipy.stats.mstats import winsorize
def winsorize_columns(data, cols, limits=(0.01, 0.01)):
    wdf = data
    for col in cols:
        wdf[col] = winsorize(wdf[col], limits=limits)
    return wdf
data = winsorize_columns(data, num_columns)
```

```python
data.to_csv('no outliers.csv', index=False)
```
